# Low Pass FIlter

```python
import numpy as np
from scipy.signal import butter, filtfilt
import matplotlib.pyplot as plt
import pandas as pd

fs= 80
nyq = 0.5*fs
order= 2
cutoff= 2
x=np.arange(fs)

dataset=pd.read_csv("/content/sample_data/1.csv")
df=pd.DataFrame(dataset)

def lowPassFilter(data, cutoff, fs, order):
  cutoffnormal=cutoff/nyq
  b, a = butter(order, cutoffnormal, btype="low")
  y= filtfilt(b, a, data)
  return y

data = df.iloc[:,1]
plt.figure(figsize=(10, 5))
plt.plot(data, label="Sinyal Asli")

temp=lowPassFilter(data,cutoff,fs,order)
plt.plot(temp, color = 'red', label="Hasil Filter")
plt.legend()
plt.show()
```
